```python
# Import necessary libraries
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer

# Download required NLTK resources (Run this once)
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# Sample document
document = """
Natural language processing (NLP) is a branch of artificial intelligence. 
It helps computers understand, interpret and manipulate human language. 
NLP draws from many disciplines, including computer science and computational linguistics.
"""
print("Original Document:\n", document)
```

```python
# Word Tokenization
tokens = word_tokenize(document)
print("Word Tokens:\n", tokens)
```

```python
# POS Tagging
pos_tags = pos_tag(tokens)
print("POS Tags:\n", pos_tags)
```

```python
# Stop Words Removal
stop_words = set(stopwords.words('english'))

# Filtering out stop words and non-alphanumeric characters (like punctuation)
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalnum()]
print("Tokens after Stop Words Removal:\n", filtered_tokens)
```

```python
# Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("Stemmed Words:\n", stemmed_words)
```

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("Lemmatized Words:\n", lemmatized_words)
```

```python
# Splitting the document into sentences to treat them as a corpus of documents
corpus = sent_tokenize(document)

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the corpus
tfidf_matrix = vectorizer.fit_transform(corpus)

# Get feature names (the vocabulary words)
feature_names = vectorizer.get_feature_names_out()

# Create a Pandas DataFrame to display the TF-IDF representation beautifully
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

print("TF-IDF Representation:\n")
display(df_tfidf)
```
