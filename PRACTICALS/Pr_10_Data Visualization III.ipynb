```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_style("whitegrid")

# Load the dataset
# Make sure 'IRIS - IRIS.csv' is in the same directory as this notebook
df = pd.read_csv('IRIS - IRIS.csv')

# Display the first 5 rows to verify
df.head()
```

```python
# Displaying information about the dataset and data types
print("Dataset Info:")
df.info()

print("\nFeature Types Summary:")
print("Numeric Features: sepal_length, sepal_width, petal_length, petal_width (Type: float64)")
print("Nominal/Categorical Feature: species (Type: object)")
```

```python
# Plotting histograms for all numeric features
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Histograms of Iris Dataset Features', fontsize=16)

sns.histplot(df['sepal_length'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Sepal Length Distribution')

sns.histplot(df['sepal_width'], kde=True, ax=axes[0, 1], color='olive')
axes[0, 1].set_title('Sepal Width Distribution')

sns.histplot(df['petal_length'], kde=True, ax=axes[1, 0], color='gold')
axes[1, 0].set_title('Petal Length Distribution')

sns.histplot(df['petal_width'], kde=True, ax=axes[1, 1], color='teal')
axes[1, 1].set_title('Petal Width Distribution')

plt.tight_layout()
plt.show()
```

```python
# Plotting box plots for all numeric features
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Box Plots of Iris Dataset Features', fontsize=16)

sns.boxplot(y=df['sepal_length'], ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Sepal Length')

sns.boxplot(y=df['sepal_width'], ax=axes[0, 1], color='olive')
axes[0, 1].set_title('Sepal Width')

sns.boxplot(y=df['petal_length'], ax=axes[1, 0], color='gold')
axes[1, 0].set_title('Petal Length')

sns.boxplot(y=df['petal_width'], ax=axes[1, 1], color='teal')
axes[1, 1].set_title('Petal Width')

plt.tight_layout()
plt.show()

# Optional: Boxplots grouped by species for better comparison
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, orient="h", palette="Set2")
plt.title("Combined Boxplot of All Features")
plt.show()
```
