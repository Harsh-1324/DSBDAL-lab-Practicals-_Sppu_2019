```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn styling for better visuals
sns.set_theme(style="whitegrid")
```

```python
# Load the dataset
df = pd.read_csv('titanic - titanic.csv')

# Display dataset info to verify 891 entries
print("--- Dataset Info ---")
df.info()

# Display the first 5 rows
print("\n--- First 5 Rows ---")
df.head()
```

```python
# Set up a figure with two subplots
plt.figure(figsize=(12, 5))

# Subplot 1: Survival by Sex
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Survived', hue='Sex', palette='pastel')
plt.title('Survival Pattern by Sex')
plt.xticks(ticks=[0, 1], labels=['Did Not Survive', 'Survived'])

# Subplot 2: Survival by Passenger Class
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Survived', hue='Pclass', palette='Set2')
plt.title('Survival Pattern by Passenger Class')
plt.xticks(ticks=[0, 1], labels=['Did Not Survive', 'Survived'])

plt.tight_layout()
plt.show()
```

```python
# Plotting the histogram for the 'Fare' column
plt.figure(figsize=(10, 6))

# Using seaborn's histplot to view the distribution, adding KDE (Kernel Density Estimate) for a smooth curve
sns.histplot(data=df, x='Fare', bins=40, kde=True, color='purple')

plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Ticket Fare ($)')
plt.ylabel('Number of Passengers')

plt.show()
```
