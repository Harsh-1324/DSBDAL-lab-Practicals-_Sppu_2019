```python
# Load the dataset
df = pd.read_csv('Iris - Iris.csv')

# Display the first 5 rows to verify the data
df.head()
```

```python
# The 'Id' column is not useful for prediction, so we drop it
df = df.drop('Id', axis=1)

# Separate features (X) and target/labels (y)
X = df.drop('Species', axis=1)
y = df['Species']

# Split the data into Training and Testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```

```python
# 1. Implement Simple Naive Bayes classification algorithm
# Initialize the Gaussian Naive Bayes classifier
nb_model = GaussianNB()

# Train the model using the training data
nb_model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = nb_model.predict(X_test)
```

```python
# 2. Compute Confusion matrix
# Since there are 3 species, this will be a 3x3 matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
# Compute TP, FP, TN, FN for each class
# TP is the diagonal elements
TP = np.diag(cm)

# FP is the sum of the respective column minus the TP
FP = cm.sum(axis=0) - TP

# FN is the sum of the respective row minus the TP
FN = cm.sum(axis=1) - TP

# TN is the total sum of the matrix minus (TP + FP + FN)
TN = cm.sum() - (FP + FN + TP)

# Mapping the values to their respective class names
classes = nb_model.classes_

for i in range(len(classes)):
    print(f"Metrics for Class: {classes[i]}")
    print(f" True Positives (TP): {TP[i]}")
    print(f" False Positives (FP): {FP[i]}")
    print(f" False Negatives (FN): {FN[i]}")
    print(f" True Negatives (TN): {TN[i]}\n")
```

```python
# Compute Accuracy, Error rate, Precision, and Recall
# We use 'macro' average to find the unweighted mean for our multi-class data
accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')

print(f"Overall Accuracy: {accuracy:.4f} (or {accuracy*100:.2f}%)")
print(f"Error Rate: {error_rate:.4f} (or {error_rate*100:.2f}%)")
print(f"Macro Precision: {precision:.4f}")
print(f"Macro Recall: {recall:.4f}")
```

```python
# Additional: Display full sklearn classification report
print("Classification Report:\n")
print(classification_report(y_test, y_pred))
```
